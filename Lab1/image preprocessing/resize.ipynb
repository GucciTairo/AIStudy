```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
# Dataset paths
train_dir = "/media/tairo/Storages/DataSetAnimals/animals/train"
test_dir = "/media/tairo/Storages/DataSetAnimals/animals/test" 
val_dir =  "/media/tairo/Storages/DataSetAnimals/animals/val"
```

```python
# Define target size
target_size = (224, 224)

def resize_images_in_place(folder_path, target_size=(224, 224)):
    for subdir, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                file_path = os.path.join(subdir, file)
                
                # Read the image
                img = cv2.imread(file_path, cv2.IMREAD_COLOR)
                if img is None:
                    print(f"Could not read {file_path}. Skipping...")
                    continue
                
                # Check if resizing is needed
                if img.shape[:2] != target_size:
                    img_resized = cv2.resize(img, target_size)
                    cv2.imwrite(file_path, img_resized)  # Overwrite the original image
                    print(f"Resized and overwritten: {file_path}")
                else:
                    print(f"Original size: {file_path}")
```

```python
# Resize images in place for each dataset
resize_images_in_place(train_dir)
resize_images_in_place(test_dir)
resize_images_in_place(val_dir)
```
